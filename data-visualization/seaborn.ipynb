```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fifa_filepath = "data/fifa.csv"
```

```python
# Read file 
fifa_data = pd.read_csv(fifa_filepath, index_col = "Date", parse_dates=True)
```

```python
fifa_data.head()
```

```python
# set the width and the height of the figure
plt.figure(figsize=(16,6))

# use line chart to show how fifa ranking evolved over time
sns.lineplot(data=fifa_data)
```

### Spotify Data

```python
spotify_filepath = "data/spotify.csv"

spotify_data = pd.read_csv(spotify_filepath, index_col = "Date", parse_dates=True)
```

### Examining the data

```python
spotify_data.head()
```

```python
# last five rows
spotify_data.tail()
```

```python
# Plotting data
sns.lineplot(data=spotify_data)
```

```python
plt.figure(figsize=(14, 6))

# add title
plt.title("Daily Global Streams of Popular Songs in 2017-2018")

# Line chart showing daily global streams of each song 
sns.lineplot(data=spotify_data)
```

### How to plot subset of the data?

```python
list(spotify_data.columns)
```

I'll plot the lines corresponding to the first two columns in the dataset.

```python
# set figure size 
plt.figure(figsize=(14, 6))

# add title
plt.title("Daily Global Streams of Popular Songs in 2017-2018")

# Line chart showing daily global streams of 'Shape of You'
sns.lineplot(data=spotify_data['Shape of You'], label="Shape of You")

# Line chart showing daily global streams of 'Despacito'
sns.lineplot(data=spotify_data['Despacito'], label="Despacito")
```
